# Plotting

## Simple interface

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 20, 1000)  # 100 evenly-spaced values from 0 to 50
y = np.sin(x)

plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlim(5, 15)
plt.ylim(-1.2, 1.2)
plt.xlabel('this is x!')
plt.ylabel('this is y!')
plt.title('My First Plot');
```

```python
plt.plot(x, y, '+r')
```

```python
plt.plot(x, y, label='test')
plt.legend()
```

## Object-oriented interface


When we called `pylab.plot` previously, there were a few things happening in the background:
 - matplotlib created a `Figure` instance, which is an object describing the plot window and its properties, and containing lists of all its elements
 - matplotlib created an `Axes` element within the figure. An axes can be thought of as a plotting window, where data can be arranged by `x` and `y` coordinates.

We're going to repeat some of the previous examples now, except we will specify these `Figure` and `Axes` instances explicitly:

```python
fig, ax = plt.subplots()
ax.plot(x, y);
```

```python
import pandas as pd

df = pd.DataFrame(zip(x, y), columns=['x', 'y'])
```

```python
fig, ax = plt.subplots()
df.plot(ax=ax)
ax.set_ylim(-1, 5)
```

```python
fig, ax = plt.subplots(2, 1)
df[['x']].plot(ax = ax[0])
df[['y']].plot(ax = ax[1])
```
